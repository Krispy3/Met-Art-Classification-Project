```python
ivory_titles = []
ivory_primaryImageSmall_urls = []
ivory_mediums = []
ivory_classifications = []
ivory_departments = []
ivory_cultures = []
ivory_periods = []
ivory_objectDates = []

attribute_containers = [ivory_titles, ivory_primaryImageSmall_urls, ivory_mediums, ivory_classifications, 
                        ivory_departments, ivory_cultures, ivory_periods, ivory_objectDates]

attributes = ['title', 'primaryImageSmall', 'medium', 'classification', 
              'department', 'culture', 'period', 'objectDate']


objects_url = 'https://collectionapi.metmuseum.org/public/collection/v1/objects/'
for objectID in ivory_objectIDs:
    r = requests.get(objects_url + str(objectID))
    for attribute_container, attribute in zip(attribute_containers, attributes):
        try:
            attribute_container.append(r.json()[attribute])
        except:
            attribute_container.append(np.nan)
#             ivory_titles.append(r.json()['title'])
#             ivory_primaryImageSmall_urls.append(r.json()['primaryImageSmall'])
#             ivory_mediums.append(r.json()['medium'])
#             ivory_classifications.append(r.json()['classification'])
#             ivory_departments.append(r.json()['department'])
#             ivory_cultures.append(r.json()['culture'])
#             ivory_periods.append(r.json()['period'])
#             ivory_objectDates.append(r.json()['objectDate'])

    time.sleep(1/80)
```

```python
# columns = ivory_objectIDs 

df = pd.DataFrame(ivory_objectIDs, columns=['Object_ID'])

df['Title'] = ivory_titles
df['URL'] = ivory_primaryImageSmall_urls
df['Medium'] = ivory_mediums
df['Classification'] = ivory_classifications
df['Department'] = ivory_departments
df['Culture'] = ivory_cultures
df['Period'] = ivory_periods
df['Object_Date'] = ivory_objectDates

df.replace(to_replace='', value=np.nan, inplace=True)

df
```
